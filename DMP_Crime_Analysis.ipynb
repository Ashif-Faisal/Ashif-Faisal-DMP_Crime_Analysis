```python
import pandas as pd
import plotly.express as px
```

```python
df=pd.read_csv('DMP_Crime_Dataset.csv', encoding = 'unicode_escape')
df.head()
```

```python
fig = px.bar(df, x='OFNS_DESC', title = "Types of Crimes")
fig.show()
```

```python
Age_wise_criminals = df.groupby(["AGE_GROUP"]).size().rename("Count").reset_index()
Age_wise_criminals.head()
```

```python
px.bar(Age_wise_criminals, x = "AGE_GROUP", y= "Count", color = "AGE_GROUP", text = "Count", title = "Criminal age wise")
```

```python
fig = px.pie(df,  names='PERP_SEX', title='Gender wise Criminals Percentage')
fig.show()
```

```python
gender_wise_criminls = px.histogram(df, x = "AGE_GROUP", color = "PERP_SEX", title = "Age wise crime with Gender")
gender_wise_criminls.show()
```

```python
map_fig = px.scatter_mapbox(df,
                            lon = df['Longitude'],
                            lat = df['Latitude'],
                            zoom = 12,
                            color = df['OFNS_DESC'],
                            size = df['KY_CD'],
                            width = 900,
                            height = 600,
                            title = 'Crime maps')
map_fig.update_layout(mapbox_style="open-street-map")
map_fig.update_layout(margin={"r": 0,"t":50,"l":0,"b":10})
map_fig.show()
```
